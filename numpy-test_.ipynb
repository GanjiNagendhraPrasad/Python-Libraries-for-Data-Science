# sum of diag using for loop

```python
import numpy as np
a=np.array([[1,7,9],[2,4,8],[3,6,5]])  # 3x3
f=0
s=0
if a.shape[0]==a.shape[1]:
  for i in a:
    f=f+i[s]
    s=s+1
  print(f)
else:
  print('doag add not possible')
```

```python
import numpy as np
import sys
a=np.array([[1,7,9,10],[2,4,8,20],[3,6,5,30],[1,2,3,4]]) # 4x4
f=0
s=0
try:
  if a.shape[0]==a.shape[1]:
    for i in a:
      f=f+i[s]
      s=s+1
    print(f)
  else:
    print('doag add not possible')
except Exception as e:
  typ,msg,lin=sys.exc_info()
  print(f'type is {typ} due to {msg} line is {lin.tb_lineno}')
```

#sum of reverse diag using for loop

```python
import sys
a=np.array([[1,7,9,10],[2,4,8,20],[3,6,5,30],[1,2,3,4]])
f=0
try:
  if a.shape[0]==a.shape[1]:

    s=-1
    for i in a:
      f=f+i[s]
      s=s-1
  else:
    print('diag add not possible')
  print(f)
except Exception as e:
  typ,msg,lin=sys.exc_info()
  print(f'type is {typ} due to {msg} line is {lin.tb_lineno}')
```

# sum of columns , and sum of that columns

```python
a=np.array([[1,7,9,10],[2,4,8,20],[3,6,5,30],[1,2,3,4]])
s=0
for i in range(len(a)):
  for j in a:
    s=s+j[i]
print(s)
```

# sum of rows , and sum of that rows

```python
a=np.array([[1,7,9,10],[2,4,8,20],[3,6,5,30],[1,2,3,4]])
s=0
for i in a:
  for j in range(len(a)):
    s=s+i[j]
print(s)
```
