```python
plt.plot(co,a,label='co-a')
plt.plot(co,b,label='co,b')
plt.plot(co,c,label='co-c')
plt.legend()
```

```python
plt.plot(co,a,label='co-a',color='r',marker='*')
plt.plot(co,b,label='co,b',color='g',marker='+')
plt.plot(co,c,label='co-c',color='b',marker='s')
plt.legend(loc=1,shadow=True)
```

#bar plot

```python
course = ['Python','Java','.NET c#','GO']
revenue=[900,1660,891,247]
```

```python
len(course)
```

```python
a=np.arange(len(course))
a
```

```python
plt.title('bar')
plt.xticks(a,course)
plt.ylabel('revenue')
plt.bar(a,revenue,label='revenue data',color='r')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.title('bar')
plt.xlabel('course')
plt.ylabel('revenue')
plt.bar(course,revenue,label='revenue data',color='r')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.title('bar')
plt.xlabel('course')
plt.ylabel('revenue')
plt.barh(course,revenue,label='revenue data',color='r')
plt.legend()
plt.show()
```

#subplots

```python
a=[2014,2015,2016,2017]
b=[4000,4500,5300,4600]
c=[2800,3000,2800,3400]

plt.plot(a , b ,color = 'r' , marker = '+',label = 'A_B')

plt.plot(a , c ,color = 'b',marker = '*',label = 'A_c')
plt.legend()
plt.show()
```

```python
a=[2014,2015,2016,2017]
b=[4000,4500,5300,4600]
c=[2800,3000,2800,3400]
plt.subplot(2,1,1)
plt.plot(a , b ,color = 'r' , marker = '+',label = 'A_B')
plt.subplot(2,1,2)
plt.plot(a , c ,color = 'b',marker = '*',label = 'A_c')
plt.legend()
plt.show()
```
