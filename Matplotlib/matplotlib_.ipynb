scatterplot

```python
import matplotlib.pyplot as plt

india=[1,5,6,2,7,4]
usa=[5,4,1,3,2,6]

plt.figure(figsize=(5,3))  #w, h
plt.title('scatterplot')
plt.xlabel('india')
plt.ylabel('usa')

plt.scatter(x=india,y=usa,color='r',marker='*')
plt.show()
```

line chart

```python
india=[1,5,6,2,7,4]
usa=[5,4,1,3,2,6]

plt.figure(figsize=(5,3))
plt.title('line chart')
plt.xlabel('india')
plt.ylabel('usa')

plt.plot(india,usa,color='r',marker='*')
plt.show()
```

```python
india=[1,5,6,2,7,4]
usa=[5,4,1,3,2,6]
uk=[6,3,5,4,1,7]

plt.figure(figsize=(5,3))
plt.title('line chart')
plt.xlabel('india')
plt.ylabel('usa')

plt.plot(india,usa,color='r',marker='*')
plt.plot(india,uk,color='g',marker='*')

plt.show()
```

```python
india=[1,5,6,2,7,4]
usa=[5,4,1,3,2,6]
uk=[6,3,5,4,1,7]

plt.figure(figsize=(5,3))
plt.title('line chart')
plt.xlabel('india')
plt.ylabel('usa')

plt.plot(india,usa,marker='*',label='india-usa')
plt.plot(india,marker='*',label='india-uk')

plt.legend()
plt.show()
```

subplot

```python
plt.figure(figsize=(8,3))  # same size for 2 boxes

plt.subplot(1,2,1)  # row wise boxes
plt.title('india-usa')
plt.plot(india,usa,color='r',marker='*')

plt.subplot(1,2,2)
plt.title('india-uk')
plt.plot(india,uk,color='r',marker='*')

plt.show()
```

```python
plt.figure(figsize=(5,5))

plt.subplot(2,1,1)  # column wise boxes
plt.title('india-usa')
plt.plot(india,usa,color='r',marker='*')

plt.figure(figsize=(10,5))  #new size for new box

plt.subplot(2,1,2)
plt.title('india-uk')
plt.plot(india,uk,color='r',marker='*')

plt.show()
```

histogram

```python
age = [1,4,7,9,3,11,14,21,27,29,31,32,35,39]

plt.figure(figsize=(5,3))

plt.hist(x=age,color='r')
plt.show()
```

```python
age = [1,4,7,9,3,11,14,21,27,29,31,32,35,39]

plt.figure(figsize=(5,3))

plt.hist(x=age,color='b',bins=4)
plt.show()
```

bar chart

```python
#vertical bar chart
company_names = ['Google','Meta','MSFT','IBM']
revenue = [1200 , 1400,1000,500]
plt.figure(figsize=(5,3))
plt.bar(company_names,revenue,color=['r','g','y','black'])
plt.show()
```

```python
#horizontal bar chart
company_names = ['Google','Meta','MSFT','IBM']
revenue = [1200 , 1400,1000,500]
plt.figure(figsize=(5,3))
plt.barh(company_names,revenue,color=['r','g','y','black'])
plt.show()
```

pie chart

```python
company_names = ['Google','Meta','MSFT','IBM']
revenue = [1200 , 1400,1000,500]

plt.figure(figsize=(5,3))
plt.pie(revenue,labels=company_names)
plt.show()
```

```python
company_names = ['Google','Meta','MSFT','IBM']
revenue = [1200 , 1400,1000,500]

plt.figure(figsize=(5,3))
plt.pie(revenue,labels=company_names,shadow=True,autopct='%.2f')
plt.show()
```

```python
company_names = ['Google','Meta','MSFT','IBM']
revenue = [1200 , 1400,1000,500]

plt.figure(figsize=(5,3))
plt.pie(revenue,labels=company_names,shadow=True,autopct='%.2f',explode=[0,0,0.2,0])
plt.show()
```
